```python
trump=trump.loc[:,['created_at', 'user_screen_name', 'is_quote', 'retweeted','user_followers_count', 'user_friends_count', 'compound']]
```

```python
trump['user_followers_count'].mean()
```

```python
trump['user_friends_count'].mean()
```

```python
temp_list=[]
n_value=[]
compound_list=trump['compound'].to_list()
followers_list=trump['user_followers_count'].to_list()
following_list=trump['user_friends_count'].to_list()
i=0
while i<len(followers_list):
    temp=(followers_list[i])-(following_list[i])

    if temp>=10:
        a=np.log10(temp)
    else:
        a=1
    
    b=float(compound_list[i])
    temp_list.append((a*b))
    n_value.append(a)

    i=i+1
```

```python
trump.head()
```

```python
trump['follower_multiplier']=n_value

```

```python
trump['follower_multiplier_2']=[i-1 for i in n_value]
trump.head()
```

```python
trump['follower_multiplier'].mean()
```

```python
i=0
sentiment_value_1_list=[]
sentiment_value_2_list=[]
while i<len(n_value):
    sentiment_value_1_list.append(compound_list[i] * n_value[i])
    sentiment_value_2_list.append(compound_list[i] * ((n_value[i])-1))
    
    
    i=i+1

trump['sentiment_value_1']=sentiment_value_1_list
trump['sentiment_value_2']=sentiment_value_2_list
```

```python
trump['created_at']=pd.to_datetime(trump['created_at'])



date_list=list(set(trump['created_at']))
date_list.sort()
```

```python
trump.set_index('created_at',inplace=True)
```

```python
date_list[:-25]
```

```python

trump=trump.loc[:'2020-11-03']
date_list=date_list[:-25]
#making stuff into list
trump_daily_sentiment_value_1=[]
trump_daily_sentiment_value_2=[]
for i in date_list:
    trump_daily_sentiment_value_1.append(trump.loc[i]['sentiment_value_1'].mean())
    trump_daily_sentiment_value_2.append(trump.loc[i]['sentiment_value_2'].mean())
   
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# prepping stuff for graphs 
JBPoll = mpatches.Patch(color='blue', label='Joe Biden Polls')
DTPoll = mpatches.Patch(color='red', label='Donald Trump Polls')
DTWPoll = mpatches.Patch(color='green', label='Trump Weighted Polling average')
JBWPoll = mpatches.Patch(color='magenta', label='Biden Weighted Polling average')
DiffPoll=mpatches.Patch(color='black', label='% difference between Biden & Trump')
yticks=[-.15,-.1,-.05,0,.05,.1]
grid_alpha=.5
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,14))

ax.plot(date_list,trump_daily_sentiment_value_1,label='trump_daily_sentiment_value_1',c="firebrick")
ax.plot(date_list,trump_daily_sentiment_value_2,label='trump_daily_sentiment_value_2',c="cornflowerblue")

ax.set_yticks(yticks)

ax.set_xlabel("Date")
ax.set_ylabel("Sentiment Analysis result")
ax.axhline(-.15, color='black',ls='--',alpha=grid_alpha)
ax.axhline(-.10, color='black',ls='--',alpha=grid_alpha)
ax.axhline(-.05, color='black',ls='--',alpha=grid_alpha)
ax.axhline(0, color='black',ls='--',alpha=grid_alpha)
ax.axhline(.05, color='black',ls='--',alpha=grid_alpha)
ax.axhline(.10, color='black',ls='--',alpha=grid_alpha)

ax.legend()
plt.show()
```
