```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_test=pd.read_csv('D:/Final Project Econ/trump sentiment and tweets.csv',nrows=100)
df=pd.read_csv('D:/Final Project Econ/trump sentiment and tweets.csv')
```

```python
df
```

```python
day_of_week=pd.to_datetime(df['created_at'])
day_of_week=day_of_week.dt.dayofweek.to_list()
df['day']=day_of_week
```

```python
week=pd.to_datetime(df['created_at']).dt.isocalendar()
df['week']=week['week']
```

```python
df
```

```python
df.set_index(['week','day'],inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df.loc[37]
```

```python
df
```

```python
df2=df.copy()
df2=df2[(abs(df2['compound'])>.025)]
```

```python
df.loc[36,3]['compound']
```

```python
df2.loc[36,3]['compound']
```

```python
mean_val=[]
mean_val_adj=[]
for week in range(36,49):

    temp=[]
    temp2=[]
    if week==36:
        for day in range(1,7):
            temp.append((df.loc[week,day]['compound'].mean(),day))
            temp2.append((df2.loc[week,day]['compound'].mean(),day))
            if day==6:
                mean_val.append(temp)
                mean_val_adj.append(temp2)
            
    elif week==48:
        for day in range(6):
            temp.append((df.loc[week,day]['compound'].mean(),day))
            temp2.append((df2.loc[week,day]['compound'].mean(),day))
            if day==5:
                mean_val.append(temp)
                mean_val_adj.append(temp2)
    else:
        for day in range(7):
            temp.append((df.loc[week,day]['compound'].mean(),day))
            temp2.append((df2.loc[week,day]['compound'].mean(),day))

            if day==6:
                mean_val.append(temp)
                mean_val_adj.append(temp2)

    
```

```python
vals=[]
days=[]


i=0
while i<len(mean_val):
    vals.extend([a for a,b in mean_val[i]])
    days.extend([b for a,b in mean_val[i]])
    i=i+1
```

```python
color_list=['#ff0012','#ff3800','#fc5500','#f76c00','#ef8000','#e59400','#d8a500','#c9b600','#b7c600','#a2d500','#88e300','#63f100','#0bfe16']
```

```python
fig, ax=plt.subplots(2,1,figsize=(13,10))
i=0

while i<len(mean_val):
    y,x=list(map(list, zip(*mean_val[i])))
    b,a=list(map(list,zip(*mean_val_adj[i])))
    ax[1].plot(a,b,label=f'week {36+i}',color=color_list[i],linewidth=2)
    ax[0].plot(x,y,label=f'week {36+i}',color=color_list[i],linewidth=2)
    

    i=i+1

ax[0].set_title('un-adjusted data')
ax[1].set_title('adjusted data')
plt.legend(loc='upper center', bbox_to_anchor=(1.04, 1.7),
          ncol=1, fancybox=True, shadow=True)
plt.show()
```
