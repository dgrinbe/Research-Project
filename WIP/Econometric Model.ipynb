```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
sentiment_t=pd.read_csv('D:/Final Project Econ/trump sentiment and tweets.csv')
polling_t=pd.read_csv('D:/Final Project Econ/trump polling data.csv')
```

```python
t_weighted_avg=pd.read_csv('D:/Final Project Econ/trump_weighted_avg.csv',parse_dates=True)
t_weighted_avg.columns=['date','weighted average']
t_weighted_avg['date']=pd.to_datetime(t_weighted_avg['date'])
t_weighted_avg.set_index('date',inplace=True)
t_weighted_avg

sentiment_t['created_at']=pd.to_datetime(sentiment_t['created_at'])
sentiment_t.set_index('created_at',inplace=True)
date_list=t_weighted_avg.index.to_list()
sentiment_t=sentiment_t.loc[date_list]
```

```python
weighted_avgs=[]
for i in date_list:
    length=len(sentiment_t['text'].loc[i])
    temp=[t_weighted_avg.loc[i].sum()]*length
    weighted_avgs.extend(temp)
sentiment_t['polling_avg']=weighted_avgs
```

```python
econometric_model=sentiment_t[['user_followers_count', 'user_friends_count','is_quote','user_verified','polling_avg','compound']]

econometric_model['log_followers']=np.log(econometric_model['user_followers_count'].to_list())
econometric_model['log_following']=np.log(econometric_model['user_friends_count'].to_list())
econometric_model['verified_eff']=np.multiply(np.array(econometric_model['compound'].to_list()),np.array(econometric_model['user_verified'].to_list()))

econometric_model
```

```python
x1=econometric_model[['user_followers_count', 'user_friends_count','is_quote','user_verified','polling_avg',]]
y1=econometric_model['compound']
x1 = sm.add_constant(x1) # adding a constant
 
model1 = sm.OLS(y1, x1).fit()
predictions1 = model1.predict(x1) 
 
print_model1 = model1.summary()
print(print_model1)
```
